```python
import os
import numpy as np
from pcapfile import savefile
from pcapfile.protocols.linklayer import ethernet
from pcapfile.protocols.network import ip
import binascii

for pcap in os.listdir('.'):
    if(len(pcap.split('.')) != 2 ):
        continue
    if(pcap.split('.')[1] != 'pcap' ):
        continue
    savepath = pcap.split('.')[0]
    if not os.path.isdir(savepath):
        os.mkdir(savepath)
    testcap = open(pcap, 'rb')
    capfile = savefile.load_savefile(testcap,layers=1, verbose=True)
    a = []
    for line in capfile.packets:
        print(line.packet.payload)
        a.append(line.packet.payload)
        
    print (len(a))
    a = np.array(a)
    outfile = savepath + '/' + 'payload.npy'
    np.save(outfile, a)
    
        
```
